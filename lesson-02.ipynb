#### `While` loop

```python
count = 1
while count <= 5: #condition to keep the loop running
    print(count)
    count += 1 #increment count by 1 in each interation. count = count +1
```

#### 2. `range`

```python
for i in range(5): #range(5) generates numbers from 0 to 4
    print(i)
```

###### `range(start, stop)`

```python
for i in range(2, 6):
    print(i)
```

#### `range(start, stop, step)`

```python
for i in range(0, 11, 2): #range(start, stop, step)
    print(i)
```

#### Counting Backwards with `range`

```python
for i in range(10, 0, -1): #range(start, stop, step)
    print(i)
```

###### `Using loops with strings`

```python
name = "Python"
for letter in name:
    print(letter)
```

#### `list`

```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

#### `break`

```python
for i in range(10): # this loop will run indefinitely because there is no condition to stop it
    if i == 5: #when i is 5, we want to stop the loop
        break
    print(i)
```

#### `Continue`

```python
for i in range(5):
    if i == 2:
        continue

    print(i)
```

#### `Summing numbers whith for loop`

```python
total = 0

for i in range(1, 6):
    total += i
print(total)
```

#### `nested loops` a loop inside another loop`

```python
for row in range(3):
    for col in range(3):
        print(f'row: {row}, col : {col}')
```

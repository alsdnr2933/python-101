#### 5. Basic Math

```python
a = 11
b = 5

print(a + b) # Addition
print(a - b) # Subtraction
print(a * b) # Multiplication 
print(a / b) # Division (result is a float)
print(a // b) # Floor division (result is an integer)
print(a % b)  # Modulo (remainder of the division)
print(a ** b) # Exponentiation (a raised to the power of b)
```

#### 6. Comparison Operators

| Operator | Meaning                |
| -------- | --------------------   |
| ==       | Equal                  |
| !=       | Not Equal              |
| >        | Greater than           |
| <        | Less than              |
| >=       | Greater than or Equal  |
| <=       | Less than or Equal     |

```python
print(10 == 10)
print(10 != 5)
print(10 > 5)
print(10 < 5)
```

#### 7. Logical Operators

| Operator | Meaning                        |
| -------- | ------------------------------ |
| and      | Both conditions must be true   |
| or       | At least one True              |
| not      | Opposite                       |

```python
print(age > 18 and age < 30)
```

```python
print(10 > 20 or 10 < 20)
```

```python
print(not True)
```

#### 8. If, elif, and else

```python
score = int(input("Enter score: "))

if score >= 90:
    print("Grade A")
elif score >= 80:
    print("Grade B")
elif score >= 70:
    print("Garde C")
elif score >= 60:
    print("Grade D")
else:
    print("Grade F")
```

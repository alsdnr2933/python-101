#### Functions

```python
print('Hello!')
print('Hello!')
print('Hello!')
```

```python
def say_hello():
    print('Hello!')

say_hello()
say_hello()
say_hello()
```

#### Functions with Parameters

```python
def greet(name):
    print('Hello', name)

greet('Alice')
greet('Bob')
```

###### Multiple Parameters

```python
def introduce(name, age):
    print(f'Hello, my name is {name} and I am {age} years old.')

introduce('Alice', 20)
introduce('Bob', 25)
introduce(age=30, name='Charlie') # You can also use keyword arguments.
```

#### Function with return

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result * 2)
```

#### Function Calling Another Function

```python
def greet():
    print('Hello')

def welcome():
    greet()
    print('Welcome')

welcome()
```

#### Returning Multiple Values

```python
def calculate(a, b):

    total = a + b
    difference = a - b

    return total, difference

sum_result, diff_result = calculate(10, 5)

print(sum_result)
print(diff_result)
```

```python
def calculate(a, b):

    total = a + b
    difference = a - b

    return total, difference

sum_result, diff_result = calculate(b=10, a=5)

print(sum_result)
print(diff_result)
```

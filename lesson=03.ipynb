#### LISTS

###### 1. Accessing List Elements

```python
fruits = ['apple', 'banana', 'orange']
print(f'fruits[0]: {fruits[0]}')
print(f'fruits[1]: {fruits[1]}')
print(f'fruits[2]: {fruits[2]}')

#Negative indexing
print(f'fruits[-1]: {fruits[-1]}')

#List and for loop
print('Using for loop to iterate over fruits:')
for i in range(len(fruits)):
    print(f'fruits[{i}]: {fruits[i]}')

# changing list items
fruits = ['apple', 'banana', 'orange']

fruits[1] = "grape"

print(f'Changed fruits: {fruits}')

# adding items
fruits = ['apple', 'banana']

fruits.append('orange')

print(f'Adding fruits: {fruits}')

# removing items

fruits = ['apple', 'banana', 'orange']
fruits.remove('banana')
print(f'Removed banana: {fruits}')

# popping items
fruits = ['apple', 'banana', 'orange']
popped_fruit = fruits.pop(1)
print(f'Popped fruit: {popped_fruit}')
```

###### List Slicing and Userful List Functions

```python
numbers = [10, 20, 30, 40, 50]
print(f'numbers[1:4]: {numbers[1:4]}')
print(f'numbers[:3]: {numbers[:3]}')
print(f'numbers[2:]: {numbers[2:]}')

# sort

numbers = [50, 20, 40, 10, 30]
numbers.sort()
print(f'Sorted numbers: {numbers}')

# reverse

numbers = [50, 20, 40, 10, 30]
numbers.reverse()
print(f'Reversed numbers: {numbers}')
```

#### Tuples

```python
colors = ('red', 'green', 'blue') # Tuples are like lists but cannot be changed.
print(f'colors[0]: {colors[0]}') # Accessing tuple items is just like lists.
```

#### Sets

```python
numbers = {1,2,3,4} # Sets store unique items and are unorderd

numbers = {1,1,1,2,2,3} # Duplicates are removed automatically.

print(numbers)

# adding items to set
numbers = {1,2,3}
numbers.add(4)
print(f'Added 4: {numbers}')

# removing items from set
numbers = {1,2,3}
numbers.remove(2)
print(f'Removed 2: {numbers}')
```

#### Dictionaries

```python
# Dictionaries have key-value pairs and are unorderd
student = {
    'name': 'Alice',
    'age': 20,
    'grade': 'A'
}
# Dictionary keys
print(f'student keys: {list(student.keys())}')

# Dictionary values
print(f'student values: {list(student.values())}')

# Accessing dictionary items

print(f'student["name"]: {student['name']}')
print(f'student["age"]: {student['age']}')

# Adding items to dictionary

student['major'] = "computer science"
print(f'Added major: {student}')
```

###### looping through dictionaries

```python
student = {
    'name': 'Alice',
    'age': 20
}

for key, value in student.items():
    print(key, value)
```

###### Nested Collecions

```python
student = [
    {
        'name': 'Alice',
        'grade': 'A'
    },
    {
        'name': 'Bob',
        'grade': 'B'
    }
]

print(student[0]['name'])
```
